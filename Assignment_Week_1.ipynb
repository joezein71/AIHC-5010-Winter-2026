```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np
```

```python
# Generate a linearly separable dataset with 2 clusters (one per class)
X, y = make_classification(
    n_samples=200,         # Total number of samples
    n_features=2,          # Number of features (for 2D visualization)
    n_informative=2,       # All features are informative
    n_redundant=0,         # No redundant features
    n_repeated=0,          # No repeated features
    n_classes=2,           # Exactly 2 classes (clusters)
    n_clusters_per_class=1, # One cluster per class
    class_sep=2.5,         # Increased separation for clearer linear separability
    random_state=42        # for reproducibility
)

# Add a small amount of jitter to make points less perfectly aligned
jitter_scale = 0.05 # Adjust for desired spread
X_jittered = X + jitter_scale * np.random.randn(*X.shape)

print(f"Shape of features (X_jittered): {X_jittered.shape}")
print(f"Shape of labels (y): {y.shape}")
print("First 5 samples (features with jitter):\n", X_jittered[:5])
print("First 5 labels:\n", y[:5])
```

Now, let's visualize this dataset to clearly see the two linearly separable clusters.

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_jittered[:, 0], X_jittered[:, 1], c=y, cmap='coolwarm', s=50, edgecolors='k')
plt.title('Linearly Separable Dataset with 2 Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
