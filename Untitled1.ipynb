```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np
```

```python
# Generate a new large synthetic dataset with 2 clusters
X_blobs_large, y_blobs_large = make_blobs(
    n_samples=1000,         # Increased total number of samples to make it large
    n_features=2,           # Number of features (for 2D visualization)
    centers=2,              # Exactly 2 cluster centers
    cluster_std=1.0,        # Standard deviation of the clusters (controls spread)
    random_state=46         # A new random state for a different dataset
)

print(f"Shape of large features (X_blobs_large): {X_blobs_large.shape}")
print(f"Shape of large labels (y_blobs_large): {y_blobs_large.shape}")
print("First 5 samples (large features):\n", X_blobs_large[:5])
print("First 5 labels (large labels):\n", y_blobs_large[:5])
```

```python
import numpy as np

def initialize_weights_bias(n_features):
    """Step 1: Initializes the weight vector to zero and bias to zero."""
    weights = np.zeros(n_features)
    bias = 0
    return weights, bias

def predict_perceptron(X, weights, bias):
    """Calculates predictions for given features, weights, and bias."""
    linear_output = np.dot(X, weights) + bias
    # Return predictions as 0 or 1
    return np.where(linear_output >= 0, 1, 0)

def train_perceptron(
    X,
    y,
    learning_rate=1.0,
    n_iterations=1000
):
    """Trains a Perceptron model according to the specified steps."""
    n_samples, n_features = X.shape
    weights, bias = initialize_weights_bias(n_features)

    # Transform labels from {0, 1} to {-1, 1} for the Perceptron update rule
    y_transformed = np.where(y == 0, -1, 1)

    # Training loop (Step 3: iterate until no updates or max_iterations reached)
    for epoch in range(n_iterations):
        num_updates = 0
        # Step 2: For each data point x
        for idx, x_i in enumerate(X):
            # Calculate activation
            linear_output = np.dot(x_i, weights) + bias

            # Step 2a: if y * (w^T * x + b) <= 0 (misclassified)
            if y_transformed[idx] * linear_output <= 0:
                # Update weights and bias: w(new) = w(old) + yx
                weights += learning_rate * y_transformed[idx] * x_i
                bias += learning_rate * y_transformed[idx]
                num_updates += 1

        # Step 3: If no updates to the weight vector in this epoch, terminate
        if num_updates == 0:
            print(f"Converged after {epoch + 1} epochs.")
            break
    else: # This block executes if the loop completes without a 'break'
        print(f"Max iterations ({n_iterations}) reached without convergence.")

    print("Training complete.")
    print(f"Final Weights: {weights}")
    print(f"Final Bias: {bias}")

    return weights, bias
```

Now that the Perceptron functions are explicitly defined, let's train a Perceptron on the `X_blobs_large` dataset using these functions and evaluate its accuracy.

```python
# Train the Perceptron using the functions on the large dataset
initial_weights_large, initial_bias_large = initialize_weights_bias(X_blobs_large.shape[1])
final_weights_large, final_bias_large = train_perceptron(
    X_blobs_large,
    y_blobs_large,
    learning_rate=1.0,
    n_iterations=1000
)

# Make predictions on the training data
y_blobs_large_pred = predict_perceptron(X_blobs_large, final_weights_large, final_bias_large)

# Calculate accuracy
accuracy_blobs_large = np.mean(y_blobs_large_pred == y_blobs_large)
print(f"Training Accuracy on X_blobs_large (using functions): {accuracy_blobs_large * 100:.2f}%")
```

Let's visualize the `X_blobs_large` dataset with the decision boundary found by the trained Perceptron, and highlight any misclassified points.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.scatter(X_blobs_large[:, 0], X_blobs_large[:, 1], c=y_blobs_large, cmap='magma', s=40, edgecolors='k', alpha=0.7, label='True Labels')

# Plot misclassified points
misclassified_large_idx = np.where(y_blobs_large_pred != y_blobs_large)
if len(misclassified_large_idx[0]) > 0:
    plt.scatter(X_blobs_large[misclassified_large_idx, 0], X_blobs_large[misclassified_large_idx, 1],
                marker='X', s=100, color='red', label='Misclassified', edgecolors='k')

# Plot the decision boundary
if final_weights_large is not None and final_bias_large is not None:
    x_min, x_max = X_blobs_large[:, 0].min() - 1, X_blobs_large[:, 0].max() + 1
    y_min, y_max = X_blobs_large[:, 1].min() - 1, X_blobs_large[:, 1].max() + 1
    xx = np.linspace(x_min, x_max, 100)
    yy = (-final_weights_large[0] * xx - final_bias_large) / final_weights_large[1]
    plt.plot(xx, yy, 'k--', label='Decision Boundary')

plt.title('Perceptron Decision Boundary on Large Synthetic Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Implement Multi-Layer Perceptron (MLP) Class

### Subtask:
Define a Python class for a simple Multi-Layer Perceptron (MLP) including methods for initialization, forward propagation with sigmoid activation, Mean Squared Error loss calculation, backpropagation, and class prediction.


**Reasoning**:
The subtask requires defining a Python class for a simple Multi-Layer Perceptron (MLP). This code block will implement the `MLP` class with all the specified methods for initialization, forward propagation, loss calculation, backpropagation, training, and prediction.



```python
class MLP:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.1):
        # Initialize weights and biases
        self.W1 = np.random.randn(input_size, hidden_size) * 0.01
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, output_size) * 0.01
        self.b2 = np.zeros((1, output_size))
        self.learning_rate = learning_rate

    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def _sigmoid_derivative(self, x):
        # The derivative of sigmoid is sigmoid(x) * (1 - sigmoid(x))
        # Here, x is already the sigmoid output, so we use x * (1 - x)
        return x * (1 - x)

    def forward(self, X):
        # Input to hidden layer
        self.hidden_input = np.dot(X, self.W1) + self.b1
        # Output of hidden layer (after activation)
        self.hidden_output = self._sigmoid(self.hidden_input)

        # Input to output layer
        self.output_input = np.dot(self.hidden_output, self.W2) + self.b2
        # Output of output layer (after activation)
        self.output_output = self._sigmoid(self.output_input)

        return self.output_output

    def compute_loss(self, y_true, y_pred):
        # Mean Squared Error
        return np.mean((y_true - y_pred)**2)

    def backward(self, X, y_true, y_pred):
        n_samples = X.shape[0]

        # Calculate error at output layer
        output_error = y_true - y_pred
        # Calculate delta for output layer
        output_delta = output_error * self._sigmoid_derivative(self.output_output)

        # Calculate error at hidden layer
        hidden_error = np.dot(output_delta, self.W2.T)
        # Calculate delta for hidden layer
        hidden_delta = hidden_error * self._sigmoid_derivative(self.hidden_output)

        # Update weights and biases for output layer
        self.W2 += np.dot(self.hidden_output.T, output_delta) * self.learning_rate
        self.b2 += np.sum(output_delta, axis=0, keepdims=True) * self.learning_rate

        # Update weights and biases for hidden layer
        self.W1 += np.dot(X.T, hidden_delta) * self.learning_rate
        self.b1 += np.sum(hidden_delta, axis=0, keepdims=True) * self.learning_rate

    def train(self, X, y, n_iterations):
        # Ensure y is 2D for consistent operations (n_samples, 1)
        y_reshaped = y.reshape(-1, 1)

        for i in range(n_iterations):
            # Forward pass
            y_pred = self.forward(X)

            # Backward pass and update weights
            self.backward(X, y_reshaped, y_pred)

            # Print loss periodically
            if (i + 1) % 100 == 0:
                loss = self.compute_loss(y_reshaped, y_pred)
                print(f"Iteration {i+1}, Loss: {loss:.4f}")

    def predict(self, X):
        # Perform a forward pass to get probabilities
        probabilities = self.forward(X)
        # Convert probabilities to binary class labels (0 or 1)
        return (probabilities > 0.5).astype(int)

print("MLP class defined successfully.")
```
